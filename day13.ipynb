```python
###step one import pandas and numpy

import pandas as pd
import numpy as np
```

```python
#step two :load data
df=pd.read_csv('sample_pivot.csv')
df.head()
df.shape
#df.ndim
```

```python
#step three: check for missing values
df.isnull().sum()
```

```python
#step 4:using statistical methods fill in missing values
df['Units'].fillna(np.mean(df['Units']), inplace=True)
df.isnull().sum()
df.max()
df.min()
```

conditional selection

```python
#selecting single condition
df[df['Region'] == 'South']
```

```python
# selecting for multi conditions
df[(df['Region']=='South') & (df['Sales']>=800)]
df[(df['Type']=="Men's Clothing") &  (df['Units']>=15)]
```

```python
##selecting by date,first the data type for date must be change from object to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
##to select based on date
df[(df['Date']>= pd.Timestamp(2020,4,1)) & (df['Date'] <= pd.Timestamp(2020,5,31)) & (df['Region']=='East')]
```

sorting of values

```python
df.sort_values('Date',ignore_index =True,ascending = True,inplace = True)
#df.sort_values('Region',ignore_index =True,ascending = True,inplace = True)
df.head()
```

Applying method

```python
##apply can be used on numeric data types  to apply a functuion eg mean,mode
df.apply('mean' , numeric_only = True)
```

```python
df.apply('max')
```

```python
df.max()
df.apply('std',numeric_only=True)
#to check a particular column 
df['Sales'].mean()
```

groupby and apply aggregate function

```python
#groupby is used to group by a certain column before performing aggregate functions
df.groupby('Type').apply('sum',numeric_only = True)
```

pivot table

```python
#used to do  statistical analysis involving more than one column
df_pv = pd.pivot_table(df,index='Region',columns= 'Type',values = ['Units','Sales',])
df_pv
```

```python
df_pv = pd.pivot_table(df,index='Region',columns= 'Type',values = ['Units','Sales',],aggfunc='max')
df_pv
```

```python
#finding overal or total
df_pv = pd.pivot_table(df,index='Region',columns= 'Type',values = ['Units','Sales',],aggfunc='sum',margins=True,margins_name='Total')
df_pv
```

cross tabulation

```python
#checks the relationship between two categorical column using 'count'
```

```python
pd.crosstab(df['Region'],df['Type'],margins = True,margins_name='Total')
```

```python
##cross tabulaton for statistical anaysis add values and agg function
pd.crosstab(df['Region'],df['Type'],margins = True,margins_name='Total',values=df['Sales'],aggfunc='mean')
```
