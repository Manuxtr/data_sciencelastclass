```python
#question : what is the distribution of work locations (remote,hybrid,onsite) across different industries
plt.style.use('ggplot')
sns.countplot(x=df['Work_Location'],palette='bright')
plt.show()
```

```python
plt.style.use('ggplot')
ax=sns.countplot(x=df['Work_Location'],palette='bright')
for p in ax.patches:
    ax.annotate(format(p.get_height(),'.0f'),(p.get_x()+p.get_width()/2.,p.get_height()),ha='center',va='center',xytext=(0,10),textcoords='offset points')
plt.show()
```

```python
#using plotly to achieve the above
fig= px.histogram(x=df['Work_Location'],text_auto=True,color=df['Work_Location'])
fig.show()
```

```python
#how do stress level vary by work location (using seaborn)
plt.style.use('ggplot')
sns.countplot(x=df['Stress_Level'],palette='bright',hue=df['Work_Location'])
plt.show()
```

```python
fig= px.histogram(x=df['Stress_Level'],text_auto=True,color=df['Work_Location'])
fig.show()
```

```python
#to changed from stack bar to cluster or group
fig= px.histogram(x=df['Stress_Level'],text_auto=True,color=df['Work_Location'],barmode='group')
fig.show()
```

```python
#what is the  average social isolation rating for each work location
df['Social_Isolation_Rating'].unique()
```

```python
fig= px.histogram(x=df['Social_Isolation_Rating'],text_auto=True,color=df['Work_Location'],barmode='group')
fig.show()
```

```python

average_social_rating=df.groupby('Work_Location')['Social_Isolation_Rating'].mean().reset_index()
outcome=average_social_rating.sort_values('Work_Location')
outcome
```

```python
sns.barplot(x=outcome['Work_Location'],y=outcome['Social_Isolation_Rating'],palette='bright')
plt.show()
```

```python


fig=px.bar(x=outcome['Work_Location'],y=outcome['Social_Isolation_Rating'],color=outcome['Work_Location'],text=outcome['Work_Location'])
fig.show()
```
