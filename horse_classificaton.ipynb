```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import class_weight
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
numerical_data = housedf.select_dtypes(include = ['int','float'])
categorical_data = housedf.select_dtypes(include = ['object','category'])
for x in numerical_data:
    housedf[x].fillna(np.mean(housedf[x]), inplace = True,)
for x in categorical_data:
    housedf[x].fillna(housedf[x].mode()[0],inplace = True)

missing_columns_values = housedf.isnull().sum()
missing_columns_per = (housedf.isnull().sum()/ len(housedf)) * 100
total_missing_values = pd.concat([missing_columns_values, missing_columns_per], axis = 1, keys = ['Missing Values', 'Percentage'])
total_missing_values = total_missing_values.sort_values('Percentage', ascending=False)
total_missing_values.head(20)
```

```python
housedf['outcome'].value_counts()
```

```python
from sklearn.utils import resample
horse_lived=housedf[housedf['outcome']=='lived']
horse_died=housedf[housedf['outcome']=='died']
horse_euthanzed=housedf[housedf['outcome']=='euthanized']
horsedied_sam=resample(horse_died,n_samples=182)
horseuthanized_sam=resample(horse_euthanzed,n_samples=182)
tweetdf=pd.concat([horse_lived,horsedied_sam,horseuthanized_sam],ignore_index=True)
tweetdf=tweetdf.sample(frac=1)
tweetdf['outcome'].value_counts()
```

```python
encoder = LabelEncoder()
categorical_data = housedf.select_dtypes(include=['object','category'])
for x in categorical_data:
    housedf[x] = encoder.fit_transform(housedf[x])
housedf.head()

```

```python
x=housedf.drop('outcome',axis=1)
y=housedf['outcome']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model1=LogisticRegression(solver='lbfgs')
model1.fit(xtrain,ytrain)
```

```python
pred1=model1.predict(xtest)
print('\nClassification Report',classification_report(ytest,pred1))
print('\nAccuracy Report\n',accuracy_score(ytest,pred1))
print('\nPrecision Report\n',precision_score(ytest,pred1,average='weighted'))
print('\nRecall Report\n',recall_score(ytest,pred1,average='weighted'))
print('\nF1 score Report\n',f1_score(ytest,pred1,average='weighted'))
print('\nConfusion metrix Report\n',confusion_matrix(ytest,pred1))
```

```python
model2=DecisionTreeClassifier(class_weight='balanced')
model2.fit(xtrain,ytrain)
```

```python
pred2=model2.predict(xtest)
print('\nClassification Report',classification_report(ytest,pred2))
print('\nAccuracy Report\n',accuracy_score(ytest,pred2))
print('\nPrecision Report\n',precision_score(ytest,pred2,average='weighted'))
print('\nRecall Report\n',recall_score(ytest,pred2,average='weighted'))
print('\nF1 score Report\n',f1_score(ytest,pred2,average='weighted'))
print('\nConfusion metrix Report\n',confusion_matrix(ytest,pred2))
```

```python
model3=RandomForestClassifier(random_state=42,n_estimators=200)
model3.fit(xtrain,ytrain)
```

```python
pred3=model3.predict(xtest)
print('\nClassification Report',classification_report(ytest,pred3))
print('\nAccuracy Report\n',accuracy_score(ytest,pred3))
print('\nPrecision Report\n',precision_score(ytest,pred3,average='weighted'))
print('\nRecall Report\n',recall_score(ytest,pred3,average='weighted'))
print('\nF1 score Report\n',f1_score(ytest,pred3,average='weighted'))
print('\nConfusion metrix Report\n',confusion_matrix(ytest,pred3))
```
